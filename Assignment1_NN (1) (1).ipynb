

Question 1: Explain the role of weights in a neuron.
 Answer:
 In an Artificial Neural Network (ANN), weights are numerical parameters assigned to each input of a neuron.

 Role of Weights:
 They determine the importance of each input feature.

 A larger weight means the input has more influence on the output.

 A smaller weight means less influence.

 During training, weights are updated to reduce error using optimization techniques like Gradient Descent.

 Mathematical Representation:
 $$
 z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
 $$
 Where:

 w
 w = weights

 x
 x = inputs

 b
 b = bias

 Weights help the model learn patterns from data.

  Question 2: What is an activation function?
   Answer:
   An activation function decides whether a neuron should be activated or not.

   It introduces non-linearity into the model so it can learn complex relationships.

   Common Activation Functions:
   Sigmoid

   ReLU (Rectified Linear Unit)

   Tanh

   Example: Sigmoid Function
   $$
   \sigma(z) = \frac{1}{1 + e^{-z}}
   $$
    
    The sigmoid function outputs values between 0 and 1, making it useful for binary classification.

     Question 3: Define probability distribution in ML context.
      Answer:
      A probability distribution describes how probabilities are assigned to different possible outcomes of a random variable.

      In Machine Learning:

      It helps model uncertainty in data.

      It is used in algorithms like Naive Bayes and Gaussian Mixture Models.

      The total probability of all outcomes equals 1.

      Examples:
      Normal (Gaussian) Distribution

      Bernoulli Distribution

      Binomial Distribution

       Question 4: What is gradient in optimization?
        Answer:
        The gradient is the derivative (slope) of a loss function with respect to model parameters.

        In Machine Learning:

        It shows the direction of maximum increase of loss.

        Used in Gradient Descent to minimize error.

        Helps update weights efficiently.

        Update Formula:
        $$
        w = w - \eta \frac{dL}{dw}
        $$

        Where:

        - $\eta$ = learning rate  
        - $\frac{dL}{dw}$ = gradient




Question 5: Create a project using Logistic Regression with TensorFlow.


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Study hours (Input)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)

# Pass (1) or Fail (0) (Output)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1], dtype=float)
```

```python
# Build Logistic Regression Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(1,))
])

model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train Model
model.fit(X, y, epochs=50)
```

```python
# Make Prediction
prediction = model.predict(np.array([4.5]).reshape(-1, 1))
print("Prediction:", prediction)
```

```python
# Visualization
plt.scatter(X, y)
plt.xlabel("Study Hours")
plt.ylabel("Pass (1) / Fail (0)")
plt.title("Logistic Regression Example")
plt.show()
```
